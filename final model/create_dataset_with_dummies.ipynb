```python
import pandas as pd
import numpy as np
```

```python
field_types_file = r'../utilfiles/codes_after_reduction.csv'
trn_dataset_with_preprocessing_path = r'../utilfiles/train_test_val_files/train_set_noNulls.csv'
val_dataset_with_preprocessing_path = r'../utilfiles/train_test_val_files/val_set_noNulls.csv'
test_dataset_with_preprocessing_path = r'../utilfiles/train_test_val_files/test_set_noNulls.csv'
```

```python
field_types_df = pd.read_csv(field_types_file)

# Create a dictionary with field names as keys and field types as values
field_types = dict(zip(field_types_df['codes'], field_types_df['dataType']))

# Separate fields based on their types
categorical_single_fields = [field for field, dtype in field_types.items() if dtype == 'Categorical (single)']
categorical_multiple_fields = [field for field, dtype in field_types.items() if dtype == 'Categorical (multiple)']

categorical_original_fields = categorical_single_fields + categorical_multiple_fields
```

```python
train_set_df = pd.read_csv(trn_dataset_with_preprocessing_path)
```

```python

train_set_with_dummies = pd.get_dummies(train_set_df, columns=[col for col in categorical_original_fields if col in train_set_df.columns])
```

```python
val_set_df = pd.read_csv(val_dataset_with_preprocessing_path)
test_set_df = pd.read_csv(test_dataset_with_preprocessing_path)
```

```python
val_set_with_dummies = pd.get_dummies(val_set_df, columns=[col for col in categorical_original_fields if col in train_set_df.columns])
test_set_with_dummies = pd.get_dummies(test_set_df, columns=[col for col in categorical_original_fields if col in train_set_df.columns])
```

```python
train_set_with_dummies.to_csv('../utilfiles/train_test_val_files/train_set_with_dummies.csv')
```

```python
val_set_with_dummies.to_csv('../utilfiles/train_test_val_files/val_set_with_dummies.csv')
test_set_with_dummies.to_csv('../utilfiles/train_test_val_files/test_set_with_dummies.csv')
```
